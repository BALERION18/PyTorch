# Matrix operation

```python
l = torch.randint(size=(2,3),low=0,high=10)
k = torch.randint(size=(3,2),low=0,high=10)
print(l)
print(k)
```

```python
# matrix multiplication
torch.matmul(l,k)
```

```python
vectora = torch.tensor([1,2])
vectorb = torch.tensor([3,4])
```

```python
# dot product (1*3+2*4)
torch.dot(vectora,vectorb)
```

```python
# transpose(matrix , jis dim ko swap krna chahte , jis dim se krna chahte)
torch.transpose(l, 0, 1)
```

```python
m = torch.randint(size=(2,2),low=0,high=10,dtype=torch.float32)
print(m)
```

```python
# determinant
torch.linalg.det(m)
```

```python
# inverse
torch.inverse(m)
```
